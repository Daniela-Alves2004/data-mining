```python
#calcular a media da coluna bmi
media_bmi = df['bmi'].mean()
print(f"media BMI:{media_bmi}")
```

Reescala dos dados

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
print(num_cols)
```

```python
df[num_cols].head()
```

```python
# z score
# Reinitialize num_cols to ensure 'id' and 'stroke' are present if this cell is run multiple times
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# remocao colunas id e stroke
if 'id' in num_cols:
  num_cols.remove('id')
if 'stroke' in num_cols:
  num_cols.remove('stroke')
```

```python




print("antes")

display(df[num_cols].head())

for col in num_cols:
  media = df[col].mean()
  desvio= df[col].std()
  df[col] = (df[col] - media) / desvio

print("depois")
display(df[num_cols].head())
```

```python
# transformacao das colunas categoricas
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
print(cat_cols)
```

```python
for cat in cat_cols:
  print(f"{cat}: {df[cat].unique()}")
```

```python
# map (columns with up to 3 categories)
df['gender'] = df['gender'].astype(str).map({'Male':0, 'Female':1, 'Other':2})
df['ever_married'] = df['ever_married'].astype(str).map({'No':0, 'Yes':1})
df['Residence_type'] = df['Residence_type'].astype(str).map({'Rural':0, 'Urban':1})

df[['gender','ever_married','Residence_type']].head()
```

```python
dummies = pd.get_dummies(df[['work_type','smoking_status']], prefix=["work","smoke"])
dummies
```

```python
# Remover as colunas categoricas originais de df
df = df.drop(columns=['work_type','smoking_status'])

# Concatenar os dummies com o DataFrame original
df = pd.concat([df, dummies],axis=1)

df.head()
```
